# Processamento de udio
## Aula 15

* Sinais de udio so, por natureza, analgicos;
* Para representarmos de forma digital,  necessrio **amostrar** o sinal.



* O grfico azul representa a onda original;
* Os pontos em vermelho so os pontos de amostragem.

* De acordo com o [Teorema de Nyquist-Shannon](https://pt.wikipedia.org/wiki/Teorema_da_amostragem_de_Nyquist%E2%80%93Shannon), um sinal deve ser amostrado a uma frequncia pelo menos **duas vezes maior** que a frequncia mxima a ser reproduzida;
* Como a audio humana escuta ondas entre 20Hz e 20kHz, uma frequncia comum  44,1kHz, que atende o teorema de Nyquist-Shannon.



### Terminologia

* **Altura**: Classifica o som como grave ou agudo;
* **udio**: Faixa de frequncias de onda que podem ser escutadas por humanos;
* [Canal](https://www.youtube.com/watch?v=4bJ0dvAl98k): Representao de udio vindo ou indo at um certo ponto;
    * Para arquivos *stereo* (maioria), haver 2 canais;
* **Intensidade**: Volume do som;
* **Profundidade de bits (bit depth)**: Quantos bits so usados para representar uma amplitude;
    * Assim como imagens, pode ser 8-, 16-, 24- ou 32-bits.
* **Taxa de amostragem (sample rate)**: Amostras por segundo;
    * 44,1kHz  o recomendado;
* **Taxa de bits (bit rate)**: Quantos bits so transmitidos por segundo;
    * Em outras palavras, a "qualidade" do som;
    * Um valor comum para som de boa qualidade  entre 64kbps e 192kbps (96kbps+ recomendado).

### Clculo da taxa de bits

$$b_r = b_d \cdot f_s \cdot n_c$$

Na qual:
* $b_r$: Taxa de bits;
* $b_d$: Profundidade de bits;
* $f_s$: Taxa de amostragem;
* $n_c$: Nmero de canais.

## Representao de udio

* Geralmente,  representado como uma matriz MxN, na qual M  o nmero de amostras e N  o nmero de canais;
* Uma forma muito simples de representar um udio  a partir de uma senoide:
$$s(x)=\sin(2\cdot\pi\cdot f\cdot x)$$

```python
# O cdigo gera uma senide com taxa de amostragem 10000 Hz,
# frequncia de 440 Hz (A4 em notao musical),
# durando 2 segundos.

fs = 10000
f = 440
t = 2

audio = np.zeros(shape = (fs * t, 1), dtype = np.float32)

for i in range(fs * t):
    audio[i, 0] = np.sin(2 * np.pi * f * i / fs)

sd.play(audio, fs)
```

```python
# O cdigo gera uma senide com taxa de amostragem 10000 Hz,
# frequncia de 349.228 Hz no canal 0 (F4 em notao musical),
# frequncia de 440 Hz no canal 1 (A4 em notao musical),
# durando 2 segundos.

fs = 10000
f1 = 349.228
f2 = 440
t = 2

audio = np.zeros(shape = (fs * t, 2), dtype = np.float32)

for i in range(fs * t):
    audio[i, 0] = np.sin(2 * np.pi * f1 * i / fs)
    audio[i, 1] = np.sin(2 * np.pi * f2 * i / fs)

sd.play(audio, fs)
```

## Manipulao de udio

* Visto que `data`  uma matriz onde cada linha representa uma amostra para cada canal, e `fs`  a frequncia de amostragem, ou o nmero de amostras por segundo, pode-se concluir que, em 1 segundo de udio, h `fs` linhas na matriz `data`;
* Se, por exemplo, deseja-se executar um udio a partir do segundo 40, pode-se fazer:

```python
fs, data = wavfile.read("Roundabout.wav")

sd.play(data[fs * 40:, :], fs)
```

```python
sd.stop()
```

* A **amplitude**  o principal responsvel pelo volume do som que ouvimos (embora tons agudos, de frequncias mais altas, so percebidos como mais altos pelo ouvido, mesmo que estejam com a mesma amplitude);
* Iremos converter o udio para o formato `float32`, de forma a no ocorrer underflow e overflow, e normalizar para valores entre `-1` e `1` (valor mximo de um `int16`  `32767`;
* Para mudar o volume do udio, basta multiplicar pontualmente a matriz `data` por um escalar, por exemplo:

```python
parte = data[fs * 40:fs * 45, :].astype(np.float32) / 32767
```

```python
sd.play(0.1 * parte, fs)
```

* Pode-se tambm variar a **frequncia do udio** (**altura**), deixando-o mais agudo ou mais grave:

```python
sd.play(parte, fs * 2)
```

```python
sd.stop()
```
