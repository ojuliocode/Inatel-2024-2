# Aula 12 - Formatos de Imagens

## Tipos de Imagens:
* **Bitmap (ou mapa de bits):** imagens formadas por pixels, em que a cada pixel é designada uma cor.
* **Vetor:** imagens formadas a partir de vetores matemáticos.
 
| Bitmap | Imagens Vetoriais |
|-|-|
| Formado por pixels | Formado por fórmulas matemáticas |
| Aliasing | Redimensionamento de imagem sem perda de qualidade |
| Possui um maior nível de detalhamento | Incapaz de representar imagens naturais |
| Muito utilizado nas fotografias | Muito utilizado em logos e animações |

## Formatos de Imagens:

### BMP (bitmap) 

* Surgiu com o sistema operacional Windows. 
* Suporta milhões de cores e preserva todos os detalhes do arquivo original.
* Arquivos extremamente grandes, pois não utilizam compressão.
* Não permite fundo transparente (canal alpha).
* Extensões: “.bmp” ou “.dib”.

<img src="bmp.svg" width="10%" height="10%" />

### JPEG

* Surgiu em 1983 pelo grupo "Joint Photographic Experts Group".
* É um dos padrões mais populares da Internet.
* Sua taxa de compressão ajustável permite a regulagem da relação entre qualidade e tamanho da imagem.
* Geralmente, ele alcança uma proporção de compressão de 10:1, sem que a perda de qualidade fique muito evidente.
* Tamanho máximo suportado: 65.535 x 65.535 pixels
* Não possue o canal alpha.
* Muito utilizado quando o tamanho do arquivo é mais importante do que a máxima qualidade de imagem (exemplo: páginas web, blogs, e-mail, redes sociais, etc.)
* Extensões: “.jpg”, “.jpeg”, “.jpe” e “.jfif”.

<img src="jpg.png" width="10%" height="10%" />

### GIF

* Surgiu em junho de 1987.
* Foi o primeiro formato de imagens com alta taxa de compressão.
* Na época, o que chamou a atenção dos usuários foi sua tecnologia de compressão de dados sem perda de qualidade, chamada LZW (Lempel-Ziv-Welch). 
* Possui a tecnologia interlaced.
* É possível ter imagens com o fundo transparente (canal alpha).
* Possuem uma limitação de cores: apenas 256 cores (8 bits).
* Possibilita que diversas imagens sejam exibidas em sequência, gerando um clipe sem áudio.
* Extensão: “.gif”

<img src="gif.gif" />

### PNG

* Surgiu em 1996 como uma melhoria do formato GIF.
* Gera imagens de alta qualidade e tamanho razoável para os padrões da internet atual.
* Não suporta animações.
* Possui uma variação de cores infinitamente maior que a dos GIFs.
* Extensão: “.png”

<img src="png.png" width="10%" height="10%" />

### SVG

* É um formato aberto, desenvolvido pela W3C e que surgiu em 2001.
* Formato vetorial para desenhos e gráficos bidimensionais, estáticos ou animados.
* Permite que o arquivo seja ampliado sem perder qualidade ou aumentar o peso.
* É suportado por todos os navegadores web modernos e pode ser animado em HTML 5.0.
* Extensão: “.svg”

<img src="svg.png" width="10%" height="10%" />

### TIFF

* Foi criado em 1986 e adquirido pela Adobe.
* Têm pouca ou nenhuma compressão.
* É um dos melhores formatos de alta qualidade para trabalhar em softwares de manipulação de imagem, como o próprio Photoshop.
* Muito utilizado para arquivos que vão ser impressos.
* Arquivo muito pesado.
* Extensões: “.tiff” ou “.tif”.

<img src="tiff.png" width="10%" height="10%" />

### RAW

* Caracteriza as imagens capturadas diretamente do sensor das câmeras digitais, sem nenhum tipo de compressão ou tratamento.
* São arquivos muito grandes, que são utilizados exclusivamente por profissionais da fotografia.
* Extensões: “.crw”, “.cr2”, “.nrw”, “.nef”.

<img src="raw.png" width="10%" height="10%" />

### PDF

* Foi criado pelo Adobe Systems em 1993.
* São muito versáteis pois podem armazenar bitmaps, vetores, textos, podem ter diversas páginas, entre outras inúmeras funções.
* Extensão: “.pdf”
