**3 -** Crie uma função que realize a reflexão no eixo x. Essa função deve receber uma **imagem como parâmetro** e **retornar uma imagem refletida no eixo x**. Aplique essa função na imagem "mustache" aberta na questão anterior e mostre o resultado.

````python
def reflexao_x(img:np.ndarray):
    (l, c, p) = img.shape 

    img_refl = np.zeros(shape=img.shape, dtype=np.uint8)

    for i in range(l):
        for j in range(c):
            newx = j
            newy = -i 
            img_refl[newy, newx] = img[i, j]
    
    return img_refl

mustache_refl = reflexao_x(mustache)
plt.imshow(mustache_refl)
    
````

````python
not_mustache_refl = np.bitwise_not(mustache_refl)
plt.imshow(not_mustache_refl)
````

**5 -** Crie uma função que realize o escalonamento de uma imagem que receba como parâmetro uma **imagem, sx e sy**, sendo sx e sy os fatores de escalonamento e **retorne a imagem escalonada**, em seguida aplique o escalonamento na imagem "shrek" para que fique do mesmo tamanho da imagem da questão anterior. Exiba o resultado.

````python
def escalonar(img:np.ndarray, sx, sy):
    (l,c,p) = img.shape 
    ls,cs = l * sy , c * sx 
    newimg = np.zeros(shape=(ls,cs,p), dtype=np.uint8)

    for i in range(ls):
        for j in range(cs):
            newy = int(np.floor(i * (l/ls)))
            newx = int(np.floor(j * (c/cs)))
            newimg[i,j] = img[newy, newx]
    
    return newimg

print(shrek.shape)
print(not_mustache_refl.shape)

shrek_escala = escalonar(shrek, 2, 2)

plt.imshow(shrek_escala)
print(shrek_escala.shape)

````

**7 -** Crie uma função para aplicar um método grayscale de sua escolha, que receba como parâmetro uma imagem e retorne uma imagem em greyscale. Aplique essa função na imagem resultante da questão anterior e mostre o resultado.

````python
def grayscale(img:np.ndarray):
    (l,c,p) = img.shape
    newimg = np.zeros(shape=(l,c), dtype=np.uint8)
    for i in range(l):
        for j in range(c):
            (r,g,b) = (float(img[i,j,0]), float(img[i,j,1]),float(img[i,j,2]))
            newimg[i,j] = (r+g+b)/3
    
    return newimg

mascara_gs = grayscale(mascara)

plt.imshow(mascara_gs, cmap='gray')
````

**8 -** Crie uma função que realize uma rotação com interpolação, a função deve receber como parâmetro **uma imagem e o ângulo alpha com que deve ser rotacionada** e **retornar uma imagem rotacionada com o angulo alpha**. Modifique o algoritmo de rotação para que funcione se necessário. Aplique a função na imagem resultante da questão anterior **com ângulo de 60°** e mostre o resultado.

````python

def rotacaoGrayscale(img:np.ndarray, angulo):
    alpha = (angulo * 2 * np.pi)/360

    #assumindo que img é grayscale
    (l,c)=img.shape 

    ls,cs = int(l * np.sqrt(2)), int(c * np.sqrt(2))

    newimg = np.zeros(shape=(ls, cs), dtype=np.uint8)

    for i in range(ls):
        for j in range(cs):
            cx = j - (ls/2)
            cy = i - (cs/2)

            newx = int(cx * np.cos(alpha) + cy * np.sin(alpha) + l/2)
            newy = int(-cx * np.sin(alpha) + cy * np.cos(alpha) + c/2)

            if 0 <= newx < c and 0 <= newy < l:
                newimg[i,j] = img[newy, newx]
    
    return newimg

mascara_gs_rotacionada = rotacaoGrayscale(mascara_gs, 60)
plt.imshow(mascara_gs_rotacionada, cmap='gray')
````
